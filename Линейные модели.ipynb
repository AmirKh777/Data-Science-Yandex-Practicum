- –û—Ñ–æ—Ä–º–∏—Ç–µ –≤—ã–≤–æ–¥—ã, —É–∫–∞–∂–∏—Ç–µ: –∑–Ω–∞—á–µ–Ω–∏–µ R¬≤ –º–æ–¥–µ–ª–∏; –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã —Å–ª–µ–¥—É—é—Ç –∏–∑ –∞–Ω–∞–ª–∏–∑–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤; –Ω—É–∂–Ω—ã –ª–∏ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è.

–û–±–Ω–∞—Ä—É–∂–µ–Ω–∞ –ª–∏–Ω–µ–π–Ω–∞—è –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å –º–µ–∂–¥—É —ç–∫–µ –∏ —É–¥–æ–µ–º. –ù–µ–ª–∏–Ω–µ–π–Ω–∞—è —Å–≤—è–∑—å —É —Å—Ç–æ–ª–±—Ü–æ–≤ –±–µ–ª–æ–∫ %, —Å–ø–æ. –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤–≤–æ–¥–∏—Ç—å –Ω–µ –Ω—É–∂–Ω–æ. –ú—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç—å –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –º–µ–∂–¥—É —ç–∫–µ –∏ —Å–ø–æ. 

```python

# —Å–æ–∑–¥–∞–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã RANDOM_STATE
RANDOM_STATE = 42

#–¥–µ–ª–∏–º –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ –≤—ã–±–æ—Ä–∫–∏
X_1 = df_1.drop(['—É–¥–æ–π, –∫–≥'], axis=1)
y_1 = df_1['—É–¥–æ–π, –∫–≥']

X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(
    X_1, 
    y_1, 
    random_state=RANDOM_STATE
)
encoder_1 = OneHotEncoder(drop='first', sparse=False) 
cat_col_names = ['–ø–æ—Ä–æ–¥–∞', '—Ç–∏–ø –ø–∞—Å—Ç–±–∏—â–∞', '–ø–æ—Ä–æ–¥–∞ –ø–∞–ø—ã_–±—ã–∫–∞', '–≤–æ–∑—Ä–∞—Å—Ç']

X_train_ohe_1 = encoder_1.fit_transform(X_train_1[cat_col_names]) 
X_test_ohe_1 = encoder_1.transform(X_test_1[cat_col_names])

# —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π encoder_col_names –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ 
encoder_col_names = encoder_1.get_feature_names() 

# —Å–æ–∑–¥–∞–π—Ç–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∏–∑ –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ø–µ—Ä–µ–¥–∞–π—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –∏–∑ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π encoder_col_names
X_train_ohe_1 = pd.DataFrame(X_train_ohe_1, columns=encoder_col_names)
X_test_ohe_1 = pd.DataFrame(X_test_ohe_1, columns=encoder_col_names)

col_col_names = ['—ç–∫–µ (—ç–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)', '—Å—ã—Ä–æ–π –ø—Ä–æ—Ç–µ–∏–Ω, –≥', '—Å–ø–æ (—Å–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)', '–∂–∏—Ä–Ω–æ—Å—Ç—å,%', '–±–µ–ª–æ–∫,%',]

# –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –º–µ—Ç–æ–¥ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
scaler_1 = StandardScaler()

X_train_scaled_1 =  scaler_1.fit_transform(X_train_1[col_col_names])

X_test_scaled_1 = scaler_1.transform(X_test_1[col_col_names])

X_train_combined_1 = pd.DataFrame(X_train_ohe_1, columns=encoder_col_names).join(
    pd.DataFrame(X_train_scaled_1, columns=col_col_names)
)


X_test_combined_1 = pd.DataFrame(X_test_ohe_1, columns=encoder_col_names).join(
    pd.DataFrame(X_test_scaled_1, columns=col_col_names)
)


# –°–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
model_lr_1 = LinearRegression()

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
model_lr_1.fit(X_train_combined_1, y_train_1)

# –ü—Ä–æ–≥–Ω–æ–∑ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_pred_1 = model_lr_1.predict(X_test_combined_1)


# –û—Ü–µ–Ω–∫–∞ –º–æ–¥–µ–ª–∏
r2 = r2_score(y_test_1, y_pred_1)
print(f"R^2 Score: {r2}")
```

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, —á—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å –Ω—É–∂–Ω–æ –ª–∏ —É–¥–∞–ª—è—Ç—å –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã.

```python
coefficients_1 = model_lr_1.coef_
print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:", coefficients_1)
```

–ü—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ —Å—Ç–æ–ª–±—Ü–∞ –°–ü–û, –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø–æ–º–µ–Ω—è–ª–∏—Å—å –Ω–µ —Å–∏–ª—å–Ω–æ, –∞ R2 score —É–ø–∞–ª–æ –¥–æ 0.71. –û—Å—Ç–∞–≤–∏–º —Ç–∞–∫—É—é –º–æ–¥–µ–ª—å.

```python
residuals_1 = y_test_1 - y_pred_1
```

**2. –ü–æ–¥–≥–æ—Ç–æ–≤—å—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å —É—á—ë—Ç–æ–º —Å–≤—è–∑–∏ –≤—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å —Ü–µ–ª–µ–≤—ã–º –∏ –æ–±—É—á–∏—Ç–µ –Ω–∞ –Ω–∏—Ö –≤—Ç–æ—Ä—É—é –º–æ–¥–µ–ª—å LinearRegression.**

- –ü–æ—Å–º–æ—Ç—Ä–∏—Ç–µ –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è: –Ω–∞ –Ω–∏—Ö –º–æ–∂–Ω–æ —É–≤–∏–¥–µ—Ç—å, —á—Ç–æ –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å–≤—è–∑–∞–Ω—ã —Å —Ü–µ–ª–µ–≤—ã–º –Ω–µ–ª–∏–Ω–µ–π–Ω–æ. –ò—Å–∫–ª—é—á–∏—Ç–µ –Ω–µ–ª–∏–Ω–µ–π–Ω–æ—Å—Ç—å, —á—Ç–æ–±—ã —ç—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏ –±—ã–ª–∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –í–æ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–ø–æ—Å–æ–±–æ–≤, –∫–æ—Ç–æ—Ä—ã–º–∏ —ç—Ç–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å:

    - –í—ã –æ–±–Ω–∞—Ä—É–∂–∏—Ç–µ, —á—Ç–æ –°–ü–û –∏ –£–¥–æ–π —Å–≤—è–∑–∞–Ω—ã –Ω–µ–ª–∏–Ω–µ–π–Ω–æ. –ù–∞ –∏—Ö –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞—Å—Å–µ—è–Ω–∏—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω—ã –≤ –¥–≤–∞ –∫–ª–∞—Å—Ç–µ—Ä–∞, –ø—Ä–∏—á—ë–º –µ—Å—Ç—å —Ç–∞–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –°–ü–û, –∫–æ—Ç–æ—Ä–æ–µ –º–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å –≥—Ä–∞–Ω–∏—Ü–µ–π –º–µ–∂–¥—É –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏. –ß—Ç–æ–±—ã —É—Å—Ç—Ä–∞–Ω–∏—Ç—å –Ω–µ–ª–∏–Ω–µ–π–Ω–æ—Å—Ç—å, –ø—Ä–µ–æ–±—Ä–∞–∑—É–π—Ç–µ –°–ü–û –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π –±–∏–Ω–∞—Ä–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–≤ —ç—Ç—É –≥—Ä–∞–Ω–∏—Ü—É.

    - –í—ã –æ–±–Ω–∞—Ä—É–∂–∏—Ç–µ, —á—Ç–æ –≠–ö–ï –∏ –£–¥–æ–π —Ç–∞–∫–∂–µ —Å–≤—è–∑–∞–Ω—ã –Ω–µ–ª–∏–Ω–µ–π–Ω–æ. –ù–µ–ª–∏–Ω–µ–π–Ω–æ—Å—Ç—å –º–µ–∂–¥—É –Ω–∏–º–∏ –º–æ–∂–Ω–æ –∏—Å–∫–ª—é—á–∏—Ç—å, –µ—Å–ª–∏ –≤–≤–µ—Å—Ç–∏ –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –≠–ö–ï –≤ –∫–≤–∞–¥—Ä–∞—Ç–µ.**

- –ü–æ–≤—Ç–æ—Ä–∏—Ç–µ —à–∞–≥–∏ –∏–∑ –ø—É–Ω–∫—Ç–∞ 1 –∏ —Å–æ–∑–¥–∞–π—Ç–µ –≤—Ç–æ—Ä—É—é –º–æ–¥–µ–ª—å: –ø–æ–¥–≥–æ—Ç–æ–≤—å—Ç–µ –¥–∞–Ω–Ω—ã–µ, –æ–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å, –æ—Ü–µ–Ω–∏—Ç–µ –µ—ë –∫–∞—á–µ—Å—Ç–≤–æ –∏ –Ω–∞–ø–∏—à–∏—Ç–µ –≤—ã–≤–æ–¥—ã.

```python
#—Å–Ω–∞—á–∞–ª–∞ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å–ø–æ –≤ –±–∏–Ω–∞—Ä–Ω—ã–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫.

df_1['—Å–ø–æ (—Å–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)'] = df_1['—Å–ø–æ (—Å–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)'].apply(lambda x: '–±–æ–ª—å—à–µ 0.92' if x >= 0.92 else '–º–µ–Ω—å—à–µ 0.92')
```

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã RANDOM_STATE
RANDOM_STATE = 42

#–¥–µ–ª–∏–º –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ –≤—ã–±–æ—Ä–∫–∏
X_2 = df_1.drop(['—É–¥–æ–π, –∫–≥'], axis=1)
y_2 = df_1['—É–¥–æ–π, –∫–≥']

X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(
    X_2, 
    y_2, 
    random_state=RANDOM_STATE
)
encoder_2 = OneHotEncoder(drop='first', sparse=False) 
cat_col_names_2 = ['–ø–æ—Ä–æ–¥–∞', '—Ç–∏–ø –ø–∞—Å—Ç–±–∏—â–∞', '–ø–æ—Ä–æ–¥–∞ –ø–∞–ø—ã_–±—ã–∫–∞', '–≤–æ–∑—Ä–∞—Å—Ç', '—Å–ø–æ (—Å–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)']

X_train_ohe_2 = encoder_2.fit_transform(X_train_2[cat_col_names_2]) 
X_test_ohe_2 = encoder_2.transform(X_test_2[cat_col_names_2])

# —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π encoder_col_names –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ 
encoder_col_names_2 = encoder_2.get_feature_names() 

# —Å–æ–∑–¥–∞–π—Ç–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∏–∑ –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ø–µ—Ä–µ–¥–∞–π—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –∏–∑ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π encoder_col_names
X_train_ohe_2 = pd.DataFrame(X_train_ohe_2, columns=encoder_col_names_2)
X_test_ohe_2 = pd.DataFrame(X_test_ohe_2, columns=encoder_col_names_2)

col_col_names_2 = ['—ç–∫–µ (—ç–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)', '—Å—ã—Ä–æ–π –ø—Ä–æ—Ç–µ–∏–Ω, –≥', '–∂–∏—Ä–Ω–æ—Å—Ç—å,%', '–±–µ–ª–æ–∫,%']

# –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –º–µ—Ç–æ–¥ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
scaler_2 = StandardScaler()

X_train_scaled_2 =  scaler_2.fit_transform(X_train_2[col_col_names_2])

X_test_scaled_2 = scaler_2.transform(X_test_2[col_col_names_2])

X_train_combined_2 = pd.DataFrame(X_train_ohe_2, columns=encoder_col_names_2).join(
    pd.DataFrame(X_train_scaled_2, columns=col_col_names_2)
)


X_test_combined_2 = pd.DataFrame(X_test_ohe_2, columns=encoder_col_names_2).join(
    pd.DataFrame(X_test_scaled_2, columns=col_col_names_2)
)


# –°–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
model_lr_2 = LinearRegression()

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
model_lr_2.fit(X_train_combined_2, y_train_2)

# –ü—Ä–æ–≥–Ω–æ–∑ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_pred_2 = model_lr_2.predict(X_test_combined_2)


# –û—Ü–µ–Ω–∫–∞ –º–æ–¥–µ–ª–∏
r2_2 = r2_score(y_test_2, y_pred_2)
print(f"R^2 Score: {r2_2}")
```

- –í—ã–ø–æ–ª–Ω–∏—Ç–µ –ø–æ–¥–≥–æ—Ç–æ–≤–∫—É –¥–∞–Ω–Ω—ã—Ö, –∞–Ω–∞–ª–æ–≥–∏—á–Ω—É—é —Ç–æ–π, –∫–æ—Ç–æ—Ä—É—é –¥–µ–ª–∞–ª–∏ –¥–ª—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.

R2 –æ—Ü–µ–Ω–∫–∏ –º–æ–¥–µ–ª–µ–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ: 0.79, 0.82, 0.825. –õ—É—á—à–µ–π –º–æ–¥–µ–ª—å—é –ø–æ –æ—Å—Ç–∞—Ç–∫–∞–º —è–≤–ª—è–µ—Ç—Å—è —Ç—Ä–µ—Ç—å—è –º–æ–¥–µ–ª—å. 

–û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤ –º–æ–¥–µ–ª–µ–π –ø–æ —Ç—Ä–µ–º –º–µ—Ç—Ä–∏–∫–∞–º: MSE, MAE, RMSE.

- –û—Ü–µ–Ω–∏—Ç–µ –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –º–µ—Ç—Ä–∏–∫–∏ accuracy, recall, precision.

- –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –º–∞—Ç—Ä–∏—Ü—É –æ—à–∏–±–æ–∫.

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã RANDOM_STATE
RANDOM_STATE = 42

# —Ä–∞–∑–±–∏–≤–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –≤—Ö–æ–¥–Ω—ã–µ –∏ —Ü–µ–ª–µ–≤–æ–π
X_4 = df_1.drop(columns=['–≤–∫—É—Å –º–æ–ª–æ–∫–∞'])
y_4 = df_1['–≤–∫—É—Å –º–æ–ª–æ–∫–∞']

#—Ä–∞–∑–±–∏–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—ã–µ –≤—ã–±–æ—Ä–∫–∏
X_train_4, X_test_4, y_train_4, y_test_4 = train_test_split(
    X_4, 
    y_4, 
    random_state=RANDOM_STATE
)

cat_col_names_4 = ['–ø–æ—Ä–æ–¥–∞', '—Ç–∏–ø –ø–∞—Å—Ç–±–∏—â–∞', '–ø–æ—Ä–æ–¥–∞ –ø–∞–ø—ã_–±—ã–∫–∞', 
                   '–≤–æ–∑—Ä–∞—Å—Ç', '—Å–ø–æ (—Å–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)']

encoder_4 = OneHotEncoder(drop='first', sparse=False) 

X_train_ohe_4 = encoder_4.fit_transform(X_train_4[cat_col_names_4]) 
X_test_ohe_4 = encoder_4.transform(X_test_4[cat_col_names_4])

# —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π encoder_col_names –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ 
encoder_col_names_4 = encoder_4.get_feature_names() 

# —Å–æ–∑–¥–∞–π—Ç–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∏–∑ –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ø–µ—Ä–µ–¥–∞–π—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –∏–∑ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π encoder_col_names
X_train_ohe_4 = pd.DataFrame(X_train_ohe_4, columns=encoder_col_names_4)
X_test_ohe_4 = pd.DataFrame(X_test_ohe_4, columns=encoder_col_names_4)


col_col_names_4 = ['—ç–∫–µ (—ç–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)', '—Å—ã—Ä–æ–π –ø—Ä–æ—Ç–µ–∏–Ω, –≥', '–∂–∏—Ä–Ω–æ—Å—Ç—å,%', '–±–µ–ª–æ–∫,%']

# –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –º–µ—Ç–æ–¥ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
scaler_4 = StandardScaler()

X_train_scaled_4 =  scaler_4.fit_transform(X_train_4[col_col_names_4])

X_test_scaled_4 = scaler_4.transform(X_test_4[col_col_names_4])

X_train_combined_4 = pd.DataFrame(X_train_ohe_4, columns=encoder_col_names_4).join(
    pd.DataFrame(X_train_scaled_4, columns=col_col_names_4)
)


X_test_combined_4 = pd.DataFrame(X_test_ohe_4, columns=encoder_col_names_4).join(
    pd.DataFrame(X_test_scaled_4, columns=col_col_names_4)
)


clf_4 = LogisticRegression()

clf_4 = clf_4.fit(X_train_combined_4, y_train_4)

y_pred_4 = clf_4.predict(X_test_combined_4)
```

```python
#–æ—Ü–µ–Ω–∏–≤–∞–µ–º –º–æ–¥–µ–ª—å

acc = accuracy_score(y_test_4, y_pred_4)

pre = precision_score(y_test_4, y_pred_4, pos_label='–≤–∫—É—Å–Ω–æ')

rec = recall_score(y_test_4, y_pred_4, pos_label='–≤–∫—É—Å–Ω–æ')

print(f'–û—Ü–µ–Ω–∫–∞ accuracy - {acc}, –æ—Ü–µ–Ω–∫–∞ precision - {pre}, –æ—Ü–µ–Ω–∫–∞ recall - {rec}]')
```
